In questa esercitazione abbiamo implementato una machine learning regression su dei dati con del rumore, utilizzando una Rete Neurale NN (Neural Network). Abbiamo in particolare studiato come l'abilità nel fittare dipenda dalla struttura della NN. 

### esercizio 11.1

Come primo esercizio abbiamo studiato il fit di una funzione lineare, nello specifico:
$$
f(x)=2x+1
$$

Il codice, e in particolare la struttura della rete, ci erano già stati forniti. Il nostro compito è stato pertanto studiare la dipendenza della regressione lineare già implementata al variare di alcuni suoi parametri caratteristici. Nella fattispecie, ci siamo concentrati sui valori di $N_{epochs}$ numero di epoche, $N_{train}$ numero dei punti di training e di $\sigma$, il rumore. 
Data la natura del problema, dove la regressione della NN doveva determinare solo due paramentri, la rete risulta composta da un unico neurone.

Per andare a valutare come una variazione dei sopracitati parametri andasse ad influenzare il fit, abbiamo considerato i valori dei due parametri caratterizzanti la retta ottenuta (coefficiente angolare $m$ e intercetta $q$), da confrontarsi con quelli attesi ($m=2$, $q=1$). 

Abbiamo mantenuto come regola generale un valore di $N_{valid}$ pari ad un decimo di $N_{train}$.

##### prima configurazione

- $\sigma = 0.3$
- $N_{epochs} = 100$
- $N_{train} = 200$







$m = 1.7865988$

$q = 0.9637866$

##### seconda configurazione

- $\sigma = 0.3$
- $N_{epochs} = 30$
- $N_{train} = 1000$







$m = 1.9846938$

$q = 1.0045913$

##### terza configurazione

- $\sigma = 0.8$
- $N_{epochs} = 30$
- $N_{train} = 1000$







$m = 1.9853656$

$q = 1.012328$

Dati i risultati ottenuti, possiamo dire che, a parità di rumore, sia da prediligire una configurazione come la seconda piuttosto che la prima, quindi in questo caso un maggior numero di dati di train rispetto che un maggior numero di epoche.

### esercizio 11.2

In questo esercizio abbiamo ceracto di estendere il modello dell'esercizio precedente in modo da poter fittare le seguente funzione polinomiale del terzo ordine:
$$
f(x)=4-3x-2x^2+3x^3
$$
per $x \in [-1,1]$.

Data la maggiore complessità della funzione (caratterizzata questa volta da 4 parametri differenti), non è stato più sufficiente utilizzare una NN costituita da un solo neurone. Abbiamo pertanto tentato diverse configurazioni, variando numero di layer e neuroni per layer. Abbiamo anche variato il numero di epoche $N_{epochs}$, mantenendo invece costanti $N_{train}=5000$, $N_{valid}=500$ e $\sigma = 0.1$.

Riportiamo di seguito i risultati ottenuti per tre diverse configurazioni provate. Nel primo grafico riportiamo la funzione da fittare con i punti di training generati con il rumore gaussiano $\sigma$. Nel secondo riportiamo l'andamento dell'errore all'aumentare delle epoche. Nel terzo la nostra predizione del valore della funzione applicata su dei punti di test appena generati. Nell'ultimo nuovamente la nostra predizione, questa volta anche al di fuori dell'intervallo di training.

##### prima configurazione

In questa configuazione abbiamo strutturato la rete in due laye, composti dai seguenti neuroni: 1 20 20 1.
Il numero di epoche scelto è stato 60.









##### seconda configurazione
In questo caso abbiamo aumentato aumentato i layer a discapito del numero di epoche. 
- $N_{epoch} = 40$
- 3 layer così composti: 1 20 12 7 1









##### terza configurazione

In questo caso invece abbiamo mantenuto un solo layer (1 80 1), aumentando il numero di epoche fino a 80.









Si evidenzia un miglior fitting ottenuto tramite la prima configurazione. In generale è possibile notare come i dati ottenuti al di fuori dell'intervallo di training siano decisamente incompatibili con quanto previsto dalla funzione di partenza. Inoltre, si può vedere come ci sia difficoltà a fittare in modo soddisfacente i valori relativi al lato più a destra dell'intervallo, in prossimità di $x=1$. Questo potrebbe essere dovuto alla presenza di un cambio di segno della derivata prima della funzione.
