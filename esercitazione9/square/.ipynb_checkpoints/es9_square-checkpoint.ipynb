```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
x, f = np.loadtxt("L2_square.txt", usecols=(0,1), unpack='true',  delimiter=',')


plt.plot(x,f)
plt.xlabel('#generazione')
plt.ylabel('L"')
plt.title("L2 best path")
plt.grid(True)
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
x, f = np.loadtxt("path_square.txt", usecols=(0,1), unpack='true',  delimiter=',')

plt.plot(x,f)
plt.plot(x,f, 'r^')
plt.xlabel('x')
plt.ylabel('y')
plt.title("Best Path")
plt.grid(True)
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, f, error = np.loadtxt("ave_square.txt", usecols=(0,1,2), delimiter=', ', unpack='true')
plt.errorbar(x,f ,yerr=error)
plt.xlabel('#generazione')
plt.ylabel('L2')
plt.title("valore medio L2 ")
plt.grid(True)
plt.show()
```
