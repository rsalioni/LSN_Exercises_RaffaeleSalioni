```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
x, f = np.loadtxt("L2_circle.txt", usecols=(0,1), unpack='true',  delimiter=',')


plt.plot(x,f)
plt.xlabel('#blocchi')
plt.ylabel('energia')
plt.title("valore L2")
plt.grid(True)
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
x, f = np.loadtxt("path_circle.txt", usecols=(0,1), unpack='true',  delimiter=',')

plt.plot(x,f)
plt.plot(x,f, 'r^')
plt.xlabel('#blocchi')
plt.ylabel('energia')
plt.title("valore L2")
plt.grid(True)
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, f, error = np.loadtxt("ave_circle.txt", usecols=(0,1,2), delimiter=', ', unpack='true')
plt.errorbar(x,f ,yerr=error)
plt.xlabel('#generazione')
plt.ylabel('L2')
plt.title("valore medio L2 ")
plt.grid(True)
plt.show()
```
