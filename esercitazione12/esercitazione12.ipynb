### esercizio 12.2

In questo secondo esercizio abbiamo cambiato l'architettura del codice dell'esercizio precedente al fine di ottenere una rete neurale convoluzionale CNN. Per questo tipo di rete ci sono due tipi di layer: un layer convoluzionale che computa la convoluzione degli input con una serie di filtri e dei polling layer che eseguono un coarse grain degli input.

Nel nostro caso abbiamo utilizzato una rete con la seguente struttura. L'input è ancora l'immagine $28 \times 28$, il primo layer convoluzionale ha $10$ filtri $5 \times 5$, poi un pooling con filtro $2 \times 2$. Successivamente applichiamo nuovamente una convoluzione con $16$ filtri  $5 \times 5$ e poi un altro pooling $2 \times 2$.
Infine applichiamo una rete neurale DNN composta da 2 layer rispettivamente da $80$ e $60$ unità.

Abbiamo testato questa rete sul data set di immagini $28 \times 28$ utilizzato per la rete dell'esercizio precedetnte.

Riportiamo di seguito i valori finali ottenuti dai test di Loss e Accuracy.

Test accuracy: 0.9810000061988831
