```python
import matplotlib.pyplot as plt
import numpy as np

kin = np.loadtxt('output_ekin.dat')
pot = np.loadtxt('output_epot.dat')
tot = np.loadtxt('output_etot.dat')
temp = np.loadtxt('output_temp.dat')
x = np.arange(kin.size)


plt.axes([0., 0., 3., 3.])
plt.plot(x,kin)
plt.plot(x,pot,'r')
plt.plot(x,tot,'g')
plt.plot(x,temp,'y')
plt.title("titolo")
plt.grid(True)
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

f, error = np.loadtxt("outputblocchi_0.dat", usecols=(0,1), delimiter=', ', unpack='true')
x = np.arange(f.size)


plt.axes([0., 0., 3., 3.])
plt.errorbar(x,f,yerr=error)
plt.xlabel('#blocchi')
plt.ylabel('en pot')
plt.grid(True)

plt.title("energia potenziale media - stato solido")
plt.show()
```
