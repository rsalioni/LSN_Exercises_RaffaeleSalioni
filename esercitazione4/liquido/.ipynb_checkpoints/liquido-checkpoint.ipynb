```python
import matplotlib.pyplot as plt
import numpy as np

kin = np.loadtxt('output_ekin.dat')
pot = np.loadtxt('output_epot.dat')
tot = np.loadtxt('output_etot.dat')
temp = np.loadtxt('output_temp.dat')
x = np.arange(kin.size)


plt.axes([0., 0., 3., 3.])
plt.plot(x,kin)
plt.plot(x,pot,'r')
plt.plot(x,tot,'g')
plt.plot(x,temp,'y')
plt.grid(True)
```
