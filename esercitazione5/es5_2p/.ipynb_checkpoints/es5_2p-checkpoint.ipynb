```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, f, error = np.loadtxt("rmedia2P.txt", usecols=(0,1,2), delimiter=', ', unpack='true')
plt.errorbar(x,f - 5.0 ,yerr=error)
plt.xlabel('#throws')
plt.ylabel('<r>-1/2')
plt.title("valore medio r - autostato 2p")
plt.grid(True)
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

X, Y, Z = np.loadtxt("pos2P.txt", usecols=(0,1,2), delimiter=', ', unpack='true')

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X, Y, Z, c=Z, marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, 30)
plt.show()
```
