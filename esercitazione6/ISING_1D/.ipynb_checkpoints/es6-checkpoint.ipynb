```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
x, f, error = np.loadtxt("ave.heat.final_metropolis", usecols=(0,1,2), unpack='true')
plt.errorbar(x,f,yerr=error)

points=100
T = np.linspace(0.5,2.0,num=points)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
heat=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)

plt.errorbar(x,f,yerr=error)
plt.plot(T, heat, 'b')
plt.title('Ising 1D, Heat capacity')
plt.xlabel('T')
plt.ylabel('C')


plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
points=100
T = np.linspace(0.5,2.0,num=points)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )

plt.plot(T, e, 'r')
plt.title('Ising 1D, internal energy')
plt.xlabel('T')
plt.ylabel('U/N')

x, f, error = np.loadtxt("ave.ene.final_metropolis", usecols=(0,1,2), unpack='true')
plt.errorbar(x,f,yerr=error)

plt.grid(True)

plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
x, f, error = np.loadtxt("ave.chi.final_metropolis", usecols=(0,1,2), unpack='true')
plt.errorbar(x,f,yerr=error)

X = beta*np.exp(2*beta*J)*(1-thN)/(1+thN)
plt.plot(T, X, 'r')
plt.title('Ising 1D, Susceptibility')
plt.xlabel('T')
plt.ylabel('$\chi$')

plt.grid(True)
plt.show()
```

```python
h=0.02 #external field
b = 1/T

x, f, error = np.loadtxt("ave.mag.final_metropolis", usecols=(0,1,2), unpack='true')
plt.errorbar(x,f,yerr=error)

l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)
plt.plot(T, M)
plt.title('Ising 1D, magnetization M with h = 0.02')
plt.xlabel('T')
plt.ylabel('$M$')

plt.grid(True)
plt.show()
```
