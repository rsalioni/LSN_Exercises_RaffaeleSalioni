```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

n_bins = 100
x = np.loadtxt("angoli.txt")
n, bins, patches = plt.hist(x, n_bins, range=(-4,4))

plt.xlabel('angolo')
plt.ylabel('Probability')
plt.title('Histogram')
plt.grid(True)

plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, f, error = np.loadtxt("pi.txt", usecols=(0,1,2), delimiter=', ', unpack='true')
plt.errorbar(x,f,yerr=error)
plt.xlabel('blocchi')
plt.ylabel('$<\pi>-real\pi$')
plt.grid(True)
plt.show()
```
