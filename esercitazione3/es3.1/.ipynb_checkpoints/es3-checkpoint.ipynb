```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, f, error = np.loadtxt("c1.txt", usecols=(0,1,2), delimiter=', ', unpack='true')
plt.errorbar(x,f-14.975790778311286,yerr=error)
plt.title("C Direct")
plt.xlabel('blocchi')
plt.ylabel('C - Cexp')
plt.grid(True)
plt.show()
```
